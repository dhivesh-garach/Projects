## Spotify makes new songs recommendations to users every week based on the songs that are currently saved in their playlists.

## Using the below dictionary which is a collection of playlists mantained by 'other' sample users, recommend songs to a new user.


```python
# Don't edit this cell, this create your playlist
import random,string
import numpy as np
songids=['SONGID'+str(i).zfill(2) for i in range(1,21)]
userid=[''.join('USER')+str(random.randint(10,99)).zfill(2) for i in range(100)]
playlist={i:j for i,j in zip(userid,[random.sample(songids,random.randint(4,7)) for i in range(len(userid))])}
```

```python
playlist
```

```python
#Let's say the new user has listened to these 3 songs
myplist=['SONGID01','SONGID06','SONGID16']
```

```python
#Creating a list of the songs of sample users from key-values pair.
userplist = list(playlist.values())
userplist
```

```python
# Checking if each song is present in the sample playlists and printing 0 and 1 (True or False)
list_a = []
for i in userplist:
    if "SONGID01" in i:
        list_a.append(1)
    else:
        list_a.append(0)
        
list_b = []
for j in userplist:
    if "SONGID06" in j:
        list_b.append(1)
    else:
        list_b.append(0)
        
list_c = []
for k in userplist:
    if "SONGID16" in k:
        list_c.append(1)
    else:
        list_c.append(0)
```

```python
# Adding the above results to check if which sample playlist has the most matching songs from new user.
sum_a_b_c = []
for x in range(len(list_a)):
    sum_a_b_c.append(list_a[x] + list_b[x] + list_c[x])
```

```python
#Finding out the index of the sample playlist which has most number of matching songs
for l in range(len(sum_a_b_c)):
    if sum_a_b_c[l] == 3:
        print("Below is the Index of the playlist which has 3 matching song:")
        print(l)
    elif sum_a_b_c[l] == 2:
        print("Below is the Index of the playlist which has 2 matching song:")
        print(l)
    elif sum_a_b_c[l] == 1:
        print("Below is the Index of the playlist which has 1 matching song:")
        print(l)
    else:
        pass
```

```python
# From above We can see that playlist at index 51 and 52 has most number of "new user" songs which are matching,
#Below are the matching playlists.
```

```python
first = userplist[51]
print("First list:")
print(first)
second = userplist[52]
print("Second list:")
print(second)
```

```python
#Removing the duplicate songs
first.remove('SONGID01')
first.remove('SONGID06')
first.remove('SONGID16')
second.remove('SONGID01')
second.remove('SONGID06')
second.remove('SONGID16')
```

```python
print(first)
print(second)
```

```python
#So now we can pick any 1 song to recommend to the new user from the matching sample playlist.
print("Recommended songs as per your current playlist:")
print(random.choice(first))
print(random.choice(second))
```
